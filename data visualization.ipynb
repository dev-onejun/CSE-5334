```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the data
data = pd.read_csv('nba_stats.csv')

print(data.head(5))
```

```python
features = data.iloc[:, 1:]  # Exclude 'Pos' column
labels = data[['Pos']]       # Select only 'Pos' as the label

print(features.head(5))
print(labels.head(5))
```

```python
# Convert 'Pos' to numerical values for correlation
labels_numeric = pd.get_dummies(labels['Pos'], prefix='Pos')

# Calculate correlation between each feature and each position type
correlation_matrix = features.corrwith(labels_numeric, axis=0)

# Convert correlation results into a DataFrame for easier plotting
correlation_df = pd.DataFrame(correlation_matrix)

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_df, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation between Features and Positions")
plt.xlabel("Positions")
plt.ylabel("Features")
plt.show()
```

```python
features = data.columns[1:]  # All columns except 'Pos'

# Set the plot style
sns.set(style="whitegrid")

# Loop through each feature to create individual plots
for feature in features:
    plt.figure(figsize=(10, 6))
    
    # Create a box plot for each feature by position
    sns.boxplot(x='Pos', y=feature, data=data)
    
    plt.title(f"Distribution of {feature} by Position")
    plt.show()
    
    # Uncomment below for violin plots
    # sns.violinplot(x='Pos', y=feature, data=data)
    # plt.title(f"Distribution of {feature} by Position (Violin Plot)")
    # plt.show()
```
